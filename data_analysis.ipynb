```python
neg_text = neg_data['question_text']
# 计算消极词频 共1396079词
neg_word = []
neg_counter = {}  # len 450687
i = 0
for line in neg_text:
    i += 1
    line = line.lower().strip("'\",.\n?？!(){}[]:;\\/").split(' ')
    if len(line) == 0:
        continue
    for w in line:
        w = w.strip("'\",.\n?？!(){}[]:;\\/")
        if not w in neg_word:
            neg_word.append(w)
        if not w in neg_counter:
            neg_counter[w] = 0
        else:
            neg_counter[w] += 1
    if i%10000 == 0:
        print(str(i)+'.........')
with open('data/neg_counter.json', 'w', encoding='utf8') as f:
    f.write(json.dumps(neg_counter, ensure_ascii = False))
```
